```python
from utils.alterar_dataset import pivotear_dataset_ureia, pivotear_dataset_creatinina,\
                                  concatenar_dois_datasets_por_paciente, deletar_valores_absurdos,\
                                  deletar_valores_absurdos
import pymannkendall as mk
import pandas as pd
import numpy as np
```

```python
df_ureia_15_dias = pd.read_csv("../databases/raw/Ureia15dias.csv", sep = ",", index_col = "subject_id")
df_creatinina_15_dias = pd.read_csv("../databases/raw/Creatinina15dias.csv", sep = ",", index_col = "subject_id")
```

```python
df_ureia_15_dias.reset_index(inplace = True)
df_ureia_15_dias_processed = pivotear_dataset_ureia(df_ureia_15_dias, "subject_id", "day", "UreiaMaxDia")
df_ureia_15_dias_processed
```

```python
df_creatinina_15_dias.reset_index(inplace = True)
df_creatinina_15_dias_processed = pivotear_dataset_creatinina(df_creatinina_15_dias, "subject_id", "day", "CreatininaMaxDia")
df_creatinina_15_dias_processed
```

```python
#df_ureia_15_dias_processed.set_index("subject_id", inplace=True)
df_pivoteado = df_ureia_15_dias_processed.T
for paciente in df_pivoteado.columns:
    print(f"""ID paciente: {paciente}
SumÃ¡rio teste de Mann-Kendall:
{mk.original_test(df_pivoteado[paciente])}""")
    print("=" * 50)
```

```python
#df_creatinina_15_dias_processed.set_index("subject_id", inplace=True)
df_pivoteado = df_creatinina_15_dias_processed.T
for paciente in df_pivoteado.columns:
    print(f"""ID paciente: {paciente}
SumÃ¡rio teste de Mann-Kendall:
{mk.original_test(df_pivoteado[paciente])}""")
    print("=" * 50)
```

```python
lista_trend:list[str] = []
lista_p:list[float] = []
lista_z:list[float] = []
lista_tau:list[float] = []
lista_variacao_s:list[float] = []
lista_slope:list[float] = []
lista_intercept:list[float] = []

df_ureia_15_dias_processed.reset_index(inplace = True)
df_transposto = df_ureia_15_dias_processed.drop(columns = ["subject_id"], inplace = False).T
for paciente in df_transposto.columns:
    trend, _, p, z, tau, s, var_s, slope, intercept = mk.original_test(df_transposto[paciente])
    lista_trend.append(trend)
    lista_p.append(p)
    lista_z.append(z)
    lista_tau.append(tau)
    lista_variacao_s.append(var_s)
    lista_slope.append(slope)
    lista_intercept.append(intercept)
    
valores_novo_dataset = {
    "subject_id"    : df_ureia_15_dias_processed["subject_id"],
    "tendencia"     : lista_trend,
    "p"             : lista_p,
    "z"             : lista_z,
    "Tau"           : lista_tau,
    "inclinacao"    : lista_slope,
    "intercepto"    : lista_intercept
}

df_ureia_15_dias_mann_kendall = pd.DataFrame(valores_novo_dataset)
df_ureia_15_dias_mann_kendall.head(10)
```
